# A verification challenge: `descending = (rev |> ascending)`

Let's prove this theorem with Imandra! 

Note: We'll skip loading the Imandra `stdlib` and its useful lemmas, so we can show everything from scratch.

# Defining the concepts

```ocaml
let rec ascending = function
  | x::y::xs -> x <= y && ascending (y::xs)
  | _ -> true
```

```ocaml
let rec descending = function
  | x::y::xs -> x >= y && descending (y::xs)
  | _ -> true
```

# Proving our theorem

We'll start by proving a few useful lemmas. We'll use Imandra's `[@@auto]` proof procedure for automated induction, and install our lemmas as rewrite rules.

The first, `desc_append` is really the 'key' concept of the proof.

```ocaml
lemma desc_append x y =
  x <> [] && y <> [] ==>
  descending (x @ y) = (descending x && descending y && List.hd (List.rev x) >= List.hd y)
[@@auto] [@@rw]
```

Next, we need to know two facts about `List.rev`. These are available in the `stdlib`, but we'll prove them from scratch for fun.

```ocaml
lemma rev_rev_id x = 
  List.rev (List.rev x) = x 
[@@auto] [@@rw]
```

```ocaml
lemma rev_empty x =
  (List.rev x = []) = (x = [])
[@@auto] [@@rw]
```

# Main theorem

We're now ready to prove our main theorem!

```ocaml
theorem main xs =
  ascending xs = descending (List.rev xs)
[@@auto]
```
